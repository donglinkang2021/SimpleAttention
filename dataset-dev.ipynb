```python
import numpy as np
import matplotlib.pyplot as plt
from dataset import regress_plane, regress_gaussian
```

```python
num_samples = 500
noise = 0.2

x, y, label = regress_plane(num_samples, noise)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=label, cmap=plt.cm.bwr)
plt.title('Regression Plane')
plt.colorbar()
plt.show()
```
