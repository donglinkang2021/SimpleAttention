# Datasets

Refer to the [neural network playground](https://playground.tensorflow.org/) of tensorflow, and I find this website datasets generated by typescript `dataset.ts`. So I use numpy implement the same function here.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_samples = 500
noise = 0.2
```

## Regression

```python
from dataset import regress_plane
x, y, label = regress_plane(num_samples, noise)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=label, cmap=plt.cm.bwr)
plt.title('Regression Plane')
plt.colorbar()
plt.savefig('images/regression_plane.png')
plt.show()
```

```python
from dataset import regress_gaussian
x, y, label = regress_gaussian(num_samples, noise)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=label, cmap=plt.cm.bwr)
plt.title('Regression Gaussian')
plt.colorbar()
plt.savefig('images/regression_gaussian.png')
plt.show()
```
