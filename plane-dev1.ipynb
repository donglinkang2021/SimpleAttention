```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plane
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(plane.valset.x, plane.valset.y, c=plane.valset.label, cmap=plt.cm.bwr)
plt.title('Regression Plane')
plt.colorbar()
plt.show()
```

```python
plane.model.eval()
# Plot the decision boundary
h = 0.02
x_min, x_max = plane.valset.x.min() - .05, plane.valset.x.max() + .05
y_min, y_max = plane.valset.y.min() - .05, plane.valset.y.max() + .05
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = plane.model(torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()]))
Z = Z.reshape(xx.shape)
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z.detach().numpy(), cmap=plt.cm.bwr, alpha=0.5)
plt.scatter(plane.valset.x, plane.valset.y, c=plane.valset.label, cmap=plt.cm.bwr)
plt.title('Regression Plane')
plt.colorbar()
plt.show()
```
